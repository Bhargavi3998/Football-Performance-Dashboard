```python

import dash


from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd

# Load the Excel file into a Pandas DataFrame
data = pd.read_excel('/Users/bhargavidwivedi/Documents/Job Applications/2024/Klout Machine/Book1.xlsx')

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("School Salary Dashboard"),
    
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in sorted(data['Year'].unique())],
        value=data['Year'].min(),  # Set default value to the earliest year
        clearable=False
    ),
    
    dcc.Graph(id='salary-bar-chart')
])

# Callback to update the bar chart based on the selected year
@app.callback(
    Output('salary-bar-chart', 'figure'),
    [Input('year-dropdown', 'value')]
)
def update_bar_chart(selected_year):
    # Filter data for the selected year
    filtered_data = data[data['Year'] == selected_year]
    
    # Find the school with the highest salary for each year
    result = filtered_data.groupby('School')['Salary'].max().reset_index()
    
    # Create the bar chart
    figure = {
        'data': [
            {
                'x': result['School'],
                'y': result['Salary'],
                'type': 'bar',
                'name': 'Salary'
            }
        ],
        'layout': {
            'title': f'Highest Salary by School in {selected_year}',
            'xaxis': {'title': 'School'},
            'yaxis': {'title': 'Salary'}
        }
    }
    return figure

# Run the Dash app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/Book_combined.xlsx'
df = pd.read_excel(file_path)

# List of columns to check for outliers
numeric_columns = ['Salary', 'Year', 'Pick', 'CAV']

# Z-Score Method to Detect Outliers
def detect_outliers_zscore(df, column_name, threshold=3):
    z_scores = np.abs(stats.zscore(df[column_name].dropna()))
    outliers = df[z_scores > threshold]
    return outliers

# IQR Method to Detect Outliers
def detect_outliers_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    return outliers

# Check for outliers in each numeric column
for column in numeric_columns:
    print(f"Checking for outliers in {column} using Z-Score method...")
    outliers_z = detect_outliers_zscore(df, column)
    print(outliers_z[[column]])
    
    print(f"\nChecking for outliers in {column} using IQR method...")
    outliers_iqr = detect_outliers_iqr(df, column)
    print(outliers_iqr[[column]])
    print("\n" + "="*60 + "\n")

    # Visualize using a boxplot
    plt.figure(figsize=(10, 5))
    plt.title(f'Boxplot for {column}')
    plt.boxplot(df[column].dropna())
    plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/Book_combined.xlsx'
df = pd.read_excel(file_path)

# List of columns to check for outliers
numeric_columns = ['Salary', 'Year', 'Pick', 'CAV']

# Ensure all numeric columns are truly numeric (convert any non-numeric values to NaN)
for column in numeric_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Z-Score Method to Detect Outliers
def detect_outliers_zscore(df, column_name, threshold=3):
    # Drop NaN values before calculating Z-Scores
    df_clean = df.dropna(subset=[column_name])
    z_scores = np.abs(stats.zscore(df_clean[column_name]))
    outliers = df_clean[z_scores > threshold]
    return outliers

# IQR Method to Detect Outliers
def detect_outliers_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    return outliers

# Check for outliers in each numeric column
for column in numeric_columns:
    print(f"Checking for outliers in {column} using Z-Score method...")
    outliers_z = detect_outliers_zscore(df, column)
    print(outliers_z[[column]])
    
    print(f"\nChecking for outliers in {column} using IQR method...")
    outliers_iqr = detect_outliers_iqr(df, column)
    print(outliers_iqr[[column]])
    print("\n" + "="*60 + "\n")

    # Visualize using a boxplot
    plt.figure(figsize=(10, 5))
    plt.title(f'Boxplot for {column}')
    plt.boxplot(df[column].dropna())
    plt.show()
```
